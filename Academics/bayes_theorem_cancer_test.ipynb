## ðŸ“Œ Overview
Bayesâ€™ Theorem helps us **reverse conditional probabilities**.

It answers: *If B happened, whatâ€™s the probability A caused it?*

Mathematically:

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

Where:
- **P(A|B)** is the probability of A *given* B happened
- **P(B|A)** is the probability of B *given* A happened
- **P(A)** is the prior probability of A
- **P(B)** is the total probability of B

## ðŸ§  Why It Matters
We often want to know how likely something is **after** weâ€™ve observed some evidence.

Example: A test comes back positive. What's the chance the person actually has the disease?

**Bayes Theorem lets us update our belief in A after seeing B.**

## ðŸ§ª Breast Cancer Screening Example
**Facts:**
- 1% of women have breast cancer â†’ P(Cancer) = 0.01
- If someone has cancer, 80% of tests will be positive â†’ P(Pos | Cancer) = 0.8
- If someone doesnâ€™t have cancer, 9.6% of tests will still be positive (false positive) â†’ P(Pos | No Cancer) = 0.096

### Goal: Whatâ€™s P(Cancer | Positive)?

```python
# Known values
p_cancer = 0.01
p_no_cancer = 0.99

p_pos_given_cancer = 0.8
p_pos_given_no_cancer = 0.096

# Total probability of a positive test
p_positive = (p_pos_given_cancer * p_cancer) + (p_pos_given_no_cancer * p_no_cancer)
```

```python
# Now letâ€™s calculate the actual probability of having cancer given a positive result
p_cancer_given_positive = (p_pos_given_cancer * p_cancer) / p_positive
print(f"ðŸ“Š Probability of cancer given a positive test: {p_cancer_given_positive:.4f} â†’ {p_cancer_given_positive * 100:.2f}%")
```

## ðŸ” Interpretation
**So what does this mean?**

*Only about 7.8% of women who test positive actually have cancer.*

Thatâ€™s surprising, right? Even though the test is **80% accurate**, it gives enough **false positives** to swamp the rare real cases.

This is the essence of Bayesian reasoning: you weigh the evidence **against the background probability.**

## ðŸ§® Recap of the Formula
\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A)}
\]
In our case:
- A = Cancer
- B = Positive Test
- \( \neg A \) = No Cancer
