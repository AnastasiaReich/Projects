# 🧠 Bayes' Theorem 
Mini-stat notebook on **Bayes’ Theorem**, a hit of probabilistic reasoning. Most used in real-world data science, medicine, finance, and especially machine learning.

## 📌 Overview
Bayes’ Theorem helps us **reverse conditional probabilities**.

It answers: *If B happened, what’s the probability A caused it?*

Mathematically:

\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
\]

Where:
- **P(A|B)** is the probability of A *given* B happened
- **P(B|A)** is the probability of B *given* A happened
- **P(A)** is the prior probability of A
- **P(B)** is the total probability of B

## 🧠 Why It Matters
We often want to know how likely something is **after** we’ve observed some evidence.

Example: A test comes back positive. What's the chance the person actually has the disease?

**Bayes Theorem lets us update our belief in A after seeing B.**

## 🧪 Breast Cancer Screening Example
**Facts:**
- 1% of women have breast cancer → P(Cancer) = 0.01
- If someone has cancer, 80% of tests will be positive → P(Pos | Cancer) = 0.8
- If someone doesn’t have cancer, 9.6% of tests will still be positive (false positive) → P(Pos | No Cancer) = 0.096

### Goal: What’s P(Cancer | Positive)?

```python
# Known values
p_cancer = 0.01
p_no_cancer = 0.99

p_pos_given_cancer = 0.8
p_pos_given_no_cancer = 0.096

# Total probability of a positive test
p_positive = (p_pos_given_cancer * p_cancer) + (p_pos_given_no_cancer * p_no_cancer)
```

```python
# Now let’s calculate the actual probability of having cancer given a positive result
p_cancer_given_positive = (p_pos_given_cancer * p_cancer) / p_positive
print(f"📊 Probability of cancer given a positive test: {p_cancer_given_positive:.4f} → {p_cancer_given_positive * 100:.2f}%")
```

## 🔍 Interpretation
**So what does this mean?**

*Only about 7.8% of women who test positive actually have cancer.*

That’s surprising, right? Even though the test is **80% accurate**, it gives enough **false positives** to swamp the rare real cases.

This is the essence of Bayesian reasoning: you weigh the evidence **against the background probability.**

## 🧮 Recap of the Formula
\[
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B|A) \cdot P(A) + P(B|\neg A) \cdot P(\neg A)}
\]
In our case:
- A = Cancer
- B = Positive Test
- \( \neg A \) = No Cancer

## 🌍 Real-Life Applications of Bayes’ Theorem
- **Medical Testing**: As we just saw — interpreting test results.
- **Spam Filtering**: P(Spam | Contains “Free iPhone”)?
- **Risk Assessment**: Bayesian credit scoring or insurance modeling.
- **Diagnostics**: From aircraft maintenance to predictive health alerts.
- **Machine Learning**: Naive Bayes algorithm (used in NLP, document classification).
